## Overview

In this lesson, we are going to build the static version of our team page. Along the way you will learn the **branch → edit → commit → pull request** workflow that professional developers use every day.

By the end of this lesson you will have:
- Built a team page by editing HTML directly
- Discovered why that approach breaks down at scale
- Pushed changes to GitHub using branches and Pull Requests
- Understood why separating *data* from *presentation* is a core principle of computing

This lesson introduces the **branch workflow** that will recur throughout the whole unit:

> `main` → pull → create branch → edit → commit → publish branch → open PR → merge PR → return to `main` → pull

At first blush, this workflow can feel really tedious. It can be positive if your team knows how to plan and compartmentalize tasks. If you do not plan properly, you will get stuck.

## 1 What is a Website?

Every webpage you visit in a browser is built from a text file called **HTML** (HyperText Markup Language). HTML uses *tags* — words wrapped in angle brackets — to describe what each piece of content is.

```html
<h3>Alex Rivera</h3>         ← a heading
<p>Digital Studies</p>       ← a paragraph
```

A browser reads those tags and turns them into the page you see. The *content* (names, text, links) and the *instructions* (tags) are mixed together in the same file.

Generally, HTML handles the structure of the page. As you will learn later, CSS controls styling and JavaScript controls the interactivity. These three languages, HTML, CSS, and JS are the foundation of modern web coding.

## 3 Commit, Push, and Open a Pull Request

After saving your edit, commit it on your branch and open a Pull Request so each person's work merges into `main` cleanly — without overwriting anyone else's card. This is the same workflow from [Introduction to GitHub](https://github.com/skills/introduction-to-github).

### Commit on your branch

1. Open **Source Control** (<img src="../lesson_assets/images/vscode/source-control.svg" alt="Source Control icon" width="14">).
2. Confirm only `team_static.html` appears under **Changes**.
   1. <img src="../lesson_assets/images/commit-changes-click.png" alt="Changes panel showing team_static.html" width="50%">
   2. If any other file appears under changes, click the **Discard Changes** icon (<img src="../lesson_assets/images/vscode/discard.svg" alt="Discard Changes icon" width="14">) next to the file. 
3. Click `+` next to **Changes** to stage all changes.
   1. <img src="../lesson_assets/images/staged-changes.png" alt="Staged Changes section with team_static.html" width="50%">
4. Click the ✨ sparkle button in the Commit comment box to AI-generate a commit message. You may need to press this twice as the first time initiates co-pilot.
   1. <img src="../lesson_assets/images/autocomment-commit.png" alt="AI-generated commit message in the Commit box" width="50%">
5. Click the drop down arrow next to **Commit** and select `Commit & Sync`.
   1. <img src="../lesson_assets/images/commit-and-sync.png" alt="Commit and Sync option in Commit dropdown" width="50%">

### Push your branch to GitHub

1. Click **Publish Branch** at the top of Source Control (or in the status bar).
   1. You might get a pop-up asking you what directory to push to. Select `origin`:
      1. <img src="../lesson_assets/images/commit-to-origin.png" alt="Pop-up asking where to push; select origin" width="50%">
   2. You will get a pop-up telling you `alex-card` or `yourname-card` has no remote branch. Click `Ok, don't ask again`.
      1. <img src="../lesson_assets/images/branch-dialogue-dont-ask-again.png" alt="Dialogue warning branch has no remote; click Ok don't ask again" width="50%">
      2. The next time you do this, the dialogue box won't appear.
   3. More pop-ups will appear on the bottom right.
      1. <img src="../lesson_assets/images/pop-up-after-push.png" alt="Notification pop-ups appearing after push" width="50%">
   4. To the question `Would you like to create a Pull Request...` answer `Don't show again`.
   5. To the question `Would you like Visual Studio Code to periodically run...` answer `Yes`.

✅ Your branch is now on GitHub.


### Open a Pull Request

Now we are going to merge your changes to your file back to the `main` branch.

1. Open **Source Control** (<img src="../lesson_assets/images/vscode/source-control.svg" alt="Source Control icon" width="14">).
2. Hover over the commit message area at the top of the **Changes** section — a **Create Pull Request** icon will appear. Click it.
   1. <img src="../lesson_assets/images/create-pull-request-click.png" alt="Create Pull Request icon in Source Control panel" width="50%">
3. A panel will open. Confirm the settings:
   - **Base:** `main`
   - **Compare:** `firstname-card` (your branch)
   - **Title:** e.g., `Add Alex Rivera card`
   - <img src="../lesson_assets/images/pull-request-dialogue.png" alt="Pull Request panel with base main and compare branch set" width="50%">
4. Click **Create**.

✅ Your PR is open and visible to the team.

### Merge and sync

Work through each person's PR **one at a time**:

1. A browser window will launch a GitHub Pull request.
   1. <img src="../lesson_assets/images/merge-pull-request-click.png" alt="GitHub Pull Request page with Merge pull request button" width="50%">
2. Click **Merge pull request**.
3. Github will check for potential errors. If none are made, you will be able to click on **Create Merge Commit**.
   1. <img src="../lesson_assets/images/create-merge-commit.png" alt="Create Merge Commit button on GitHub PR page" width="50%">
4. Once the Merge Commit is successful you can `delete branch`.
   1. <img src="../lesson_assets/images/delete-branch.png" alt="Delete branch button after successful merge" width="50%">
   2. A pop-up will appear at the top of your screen. Click `Ok`.
      1. <img src="../lesson_assets/images/delete-branch-pop-up.png" alt="Confirmation pop-up to delete branch" width="50%">
5. Switch back to the `main` branch in the status bar.
6. Open **Source Control** (<img src="../lesson_assets/images/vscode/source-control.svg" alt="Source Control icon" width="14">).
7. Click **Sync Changes** to download the merged commits from GitHub.
   1. <img src="../lesson_assets/images/source-control-sync.png" alt="Sync Changes button in Source Control" width="50%">
8. The Source Control should update and you should see the main branch with your name-card branch loop from one node to another. This tells you visually that you created a branch, made some changes, and then merged the branch back into the main branch.
   1. <img src="../lesson_assets/images/completed-merge.png" alt="Source Control graph showing branch merged into main" width="50%">
9. Everyone should sync their source control and check `team_static.html` — all changes should now be visible.

✅ Everyone should now see the same updated page on `main`.

> 👉 **Note:** *Always pull `main` before creating a new branch. If you skip the pull step, your new branch will be missing other people's merged changes. This causes merge conflicts and empty PRs.*

### Rules to Remember

- Work only on your own branch — never commit directly to `main`.
- Edit only your assigned member card.
- Standard workflow: `main` → pull → create branch → edit → commit → publish branch → open PR → merge PR → return to `main` → pull.

<details>
<summary><strong>⚠️ Getting errors creating a PR? (<code>Head sha can't be blank</code> / <code>No commits between main and firstname-card</code>)</strong></summary>

This means your commits accidentally went to **local `main`** instead of your feature branch. Fix it with these commands in the Terminal (`Ctrl+`` on Windows / `Cmd+`` on Mac):

```bash
# Step 1 — rescue your commits onto a new branch (run from local main)
git checkout -b firstname-card

# Step 2 — reset local main back to GitHub's version
git checkout main
git reset --hard origin/main

# Step 3 — switch back to your rescue branch and publish it
git checkout firstname-card
```

Then click **Publish Branch** in Source Control and continue with **Open a Pull Request** below.

</details>

> 💡 **Reflection:** Git records a complete history of every commit, while saving a file only keeps the most recent version. Why might this history be useful when you realize three days later that a change you made broke something?

## 5 The Problem with Hardcoding

What you just did is called **hardcoding** — writing specific values directly into a file that is supposed to be *structure*, not *content*.

Here's the core problem:

> When data and presentation are mixed together, every change to the data requires a human to find and edit code.

In a small team this is annoying. At real scale — a company website with thousands of employees, a university directory, a news article database — it becomes impossible.

The solution is **separation of concerns**: keep the data in one place, keep the presentation instructions in another, and write code that connects them.

You already use this pattern every day without thinking about it. A Google Sheets spreadsheet is *data*. The chart Google Sheets draws from it is *presentation*. If you add a row to the spreadsheet, the chart updates automatically. You didn't touch the chart.

That is exactly what we are going to do with your team page — but first, complete **Lesson 1.2** to see what happens when two people edit the same file at the same time.

## Lesson Summary

### Part 1: HTML and Websites
- **HTML** — the language that defines the structure and content of a webpage
- **Tags** — `<h3>`, `<p>`, etc. wrap content and tell the browser how to display it
- **Hardcoding** — mixing data directly into structure; works for small cases but breaks at scale

### Part 2: The Branch Workflow
- **Branch** — an isolated copy of the code where you can make changes safely
- **Commit** — a saved snapshot of your changes with a message describing what you did
- **Pull Request (PR)** — a request to merge your branch's changes into `main`
- **Merge** — combining the changes from a branch into `main`
- **Sync** — downloading the latest changes from GitHub to your local machine

### Part 3: Separation of Concerns
- Keeping *data* and *presentation* separate makes code easier to update and maintain
- Changes to data should not require editing the structure of a page
