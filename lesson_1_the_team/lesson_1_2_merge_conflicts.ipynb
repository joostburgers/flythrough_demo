## 1 What is a Merge Conflict?

A **merge conflict** occurs when two branches each contain a commit that changes the *same part of the same file* in different ways. Git cannot decide which version is correct, so it stops and asks you to choose.

This sounds scary, but it is a normal part of collaborative coding. Learning to handle conflicts confidently is one of the most valuable skills you can build.

### When do conflicts happen?

| Scenario | Conflict? |
|---|---|
| Two people edit *different* files | No |
| Two people edit *different lines* of the same file | Usually no |
| Two people edit the *same line(s)* of the same file | **Yes** |
| One person deletes a file another person edited | **Yes** |

In Lesson 1.1 everyone touched a different card, so there were no conflicts. In this lesson you will deliberately trigger one.

## 2 Create a Conflict (Pair Exercise)

This exercise works best in **pairs** — Person A and Person B. If your team has an odd number, one group of three can have Person A, Person B, and an observer who follows along on someone's screen.

### Setup — both people start from `main`

1. Make sure you are on the `main` branch (status bar, bottom left).
2. Open **Source Control** (<img src="../lesson_assets/images/vscode/source-control.svg" alt="Source Control icon" width="14">) and click **Sync Changes** to make sure you have the latest version.

### Person A — create branch and edit the page title

1. Create a new branch named `person-a-title`.
2. Open `team_static.html` in Explorer.
3. Find the `<h2>` tag near the top of the file that contains the title Meet Our Team:
   ```html
   <main class="container">
        <section class="team-intro">
            <h2>Meet Our Team</h2>
            <!-- Change this sentence to describe your group -->
            <p>This project was created by Digital Studies students at James Madison University.</p>
        </section>

   ```
4. Change it to:
   ```html
   <h2>Meet Our Team Members</h2>
   ```
5. Save, commit with a message like `Update page title to Our Team`, and publish the branch.
6. Open a Pull Request to `main` but **do not click create yet**.

### Person B — create a *different* branch and edit the *same* line

1. Stay on `main` (do **not** pull Person A's branch).
2. Create a new branch named `person-b-title`.
3. Open `team_static.html`.
4. Find the same `<h2>` tag and change it to:
   ```html
    <h2>Meet the Team</h2>
   ```
5. Save, commit with a message like `Update page title to Meet the Team`, and publish the branch.
6. Open a Pull Request to `main`. Click **Merge pull request** → **Create Merge Commit** → **Delete branch**.
7. Sync `main` to pull in changes.

✅ You now have a situation where `person-a-title` and `person-b-title` started from the same `main`, but `person-b-title` changed 'main'. Now, when a pull request is activated for `person-a-title` it will be trying to change a different version of `main` leading to a conflict.

## 3 Trigger the Conflict

1. Go to the open **Pull Request** tab for `person-a-title`.
2. Click **Create**.
GitHub will show a warning banner:

   The green **Merge pull request** button is greyed out or replaced with a **Resolve conflicts** button.
   <img src="../lesson_assets/images/branch-conflict.png" alt="Branch conflict" width="50%">


✅ You have successfully created a merge conflict. Now let's fix it.

---

## 4 Resolve the Conflict in Codespaces

Codespaces provides a visual **Merge Editor**. 

1. Click on **Resolve conflicts**.
   1. Three windows will open: 
   - The top left are **incoming** changes 
   - The top right are **current changes**  
   - The bottom window is the current state of 'main'.
2. Each merge error will be highlighted in green.
3. Click on **Accept incoming** or **Accept current** to accept that specific change.
   1. In our case **Accept incoming** - There is no real reason for this!
   2. <img src="../lesson_assets/images/accept-current-incoming.png" alt="accept current incoming">
4. Once all conflicts have been resolved, click **Complete Merge** at the bottom-right.
   1. <img src="../lesson_assets/images/complete-merge-after-conflict.png" alt="complete merge">
5. You might get a pop-up on the bottom right asking you to **Commit**. Click **Commit**.
6. The Pull Request will now revert to **Merge Pull Request**. You are free to merge as usual.
   
✅ Both PRs are merged and `main` contains a single agreed-upon title.

> 💡 **Reflection:** This conflict happened because two people edited the same line of the same file at the same time. What communication or workflow habits could your team adopt to make this situation less likely in a larger project?

## 5 Clean Up

1. Switch back to `main` in the status bar.
2. Open **Source Control** and click **Sync Changes** to pull the final merged result.
3. Open `team_static.html` and confirm there are **no conflict markers** (`<<<<<<<`, `=======`, `>>>>>>>`) left in the file.
4. Verify the `<h2>` tag contains the agreed-upon title.

✅ `main` is clean and everyone should Sync to get the latest version.

## 6 How to Prevent Most Conflicts

Conflicts are unavoidable in collaborative work, but most can be minimised with good habits:

1. **Pull `main` before branching.** If your branch starts from the latest `main`, there is less divergence to reconcile.
2. **Keep branches short-lived.** The longer a branch lives, the more `main` moves on without it. Merge early and often.
3. **Divide ownership clearly.** When possible, assign different files (or clearly separate sections of a file) to different team members.
4. **Communicate before editing shared files.** A quick message — "I'm about to edit the header" — prevents two people starting the same edit simultaneously.
5. **Merge one PR at a time.** Merge PR 1, then everyone pulls `main`, then open PR 2. Merging stacked PRs simultaneously multiplies conflicts.

## Lesson Summary

### Part 1: What is a Merge Conflict?
- A **merge conflict** occurs when two branches change the same line of the same file in different ways
- Git cannot auto-resolve the conflict and asks you to choose

### Part 2: Resolving Conflicts
- **Merge Editor** — Codespaces' visual tool for choosing between incoming and current changes
- **Accept incoming / Accept current** — pick which version of the conflicting line to keep
- **Complete Merge** — finalises the resolution and re-enables the PR merge button

### Part 3: Prevention
- Pull `main` before branching
- Keep branches short-lived and focused
- Divide file ownership clearly across team members
- Merge one PR at a time
