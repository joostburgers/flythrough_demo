## 1 Calling Functions

You have already been using **functions** — `type()`, `print()`, and `int()` are all functions. A function is a named operation: you give it input inside parentheses, and it produces output.

```python
type(a)           # tells you the data type of a
print("Hello")    # displays text
int("5309")       # converts the string "5309" to the integer 5309
```

Some functions take **keyword arguments** — named options that control exactly how the function runs. You will see this pattern constantly:

```python
pd.read_csv("data.csv", encoding="utf-8")
df.sample(n=5, random_state=42)
```

The `n=5` and `encoding="utf-8"` parts are keyword arguments. The `name=value` pattern is all you need to recognize. Generally, well-defined functions will provide documentation  will always tell you what names are available and possible values.

**A preview of Lesson 3:** In the next lesson you will use `plotly.express` to create interactive charts from Reddit data. The function call looks like this:

```python
fig = px.bar(x=universities, y=scores, title="Reddit Scores by University")
fig.show()
```

Notice the pattern: keyword arguments (`x=`, `y=`, `title=`) control what the chart displays, and the result is stored in a variable (`fig`) so you can call the `.show()` method on it. Run the cell below to see it in action.

> 👉 **Note:** *The `import plotly.express as px` line at the top of the cell loads the Plotly tool. You will learn exactly how `import` works in Lesson 2.3.*

````python
import plotly.express as px #imports the plotly library

universities = ["JMU", "UVA", "Virginia Tech", "William and Mary"] # Creates a list of data called universities that contains the values in the brackets
scores = [1204, 892, 3041, 756] # Creates a list of score values

fig = px.bar(x=universities, y=scores, title="YOUR TITLE HERE")
fig.show()
````

## 2 Methods

A **method** is a function that belongs to a specific data type and is accessed with dot notation. When you write `my_string.lower()`, you are asking the string to perform an operation on itself.

The key pattern to recognize is:

```
variable.method_name(arguments)
```

In Lesson 3 you will start working with data tables (called DataFrames in pandas). Methods appear there too — they work the same way, just applied to columns instead of individual strings:

```python
df['university'].str.lower()    # convert a whole column to lowercase
df['score'].mean()              # calculate the average of a column
```

**Common string methods — keep this as a reference:**

| Method | What it does | Example |
|---|---|---|
| `.lower()` | all lowercase | `"JMU".lower()` → `"jmu"` |
| `.upper()` | ALL UPPERCASE | `"jmu".upper()` → `"JMU"` |
| `.title()` | Title Case | `"james madison".title()` → `"James Madison"` |
| `.strip()` | remove leading/trailing spaces | `"  JMU  ".strip()` → `"JMU"` |
| `.replace("old", "new")` | swap text | `"JMU".replace("JMU", "Dukes")` → `"Dukes"` |

Methods always use **dot notation**: the variable, a dot, the method name, and parentheses.

You can also **chain methods** — call one after another — in a single line:

```python
messy_text = "  COLONIAL WILLIAMSBURG  "
clean_text = messy_text.strip().title()
# Result: "Colonial Williamsburg"
```

This is common in data cleaning: `.strip()` removes extra spaces, then `.title()` capitalizes properly. You will see chains like this throughout these lessons.

## Lesson Summary

### Part 1: Calling Functions
- **Function** — a named operation that takes input in parentheses and produces output
- **Keyword arguments** — named options passed to a function using `name=value` syntax
- **Pattern** — `function_name(argument1, keyword=value)` is the standard call format; you will see this everywhere in pandas and Plotly

### Part 2: Methods
- **Method** — a function that belongs to a specific data type, called with dot notation: `variable.method_name()`
- **Common string methods** — `.lower()`, `.upper()`, `.title()`, `.strip()`, `.replace()`
- **Method chaining** — methods can be chained in sequence: `text.strip().title()` applies both operations in one line
- **Look ahead** — DataFrames in pandas use the same dot-notation pattern; `df['column'].mean()` works exactly like `"text".lower()`
