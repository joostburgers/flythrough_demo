## 1 For Loops

One of the most common features of any programming language is **iteration**. This is when you repeat the same operation for every item in a collection. In Python this is done with a `for` loop.

A `for` loop has two parts:
- A **header** that says what to loop over
- An **indented body** that runs once for each item

The loop below goes through every university in a list and prints it in uppercase. **Read it as:** "For each `x` in `list4`, print `x` in uppercase."

You will see this pattern throughout these lessons. You do not need to write `for` loops — just recognize what they are doing when you see them.

````python
list4 = ["jmu", "uva", "virginia tech", "william and mary"]

for x in list4:
    print(x.upper())
````

## 2 Lambda

A `lambda` is a compact, one-line function. You will see it most often with `.apply()`, which runs an operation on every row of a data column:

```python
df['score'].apply(lambda x: x * 100)
```

**How to read it:** "For each value `x` in the `score` column, multiply it by 100."

Think of `lambda x:` as "a temporary variable called `x` that holds each item one at a time." You will not write lambdas in these lessons, but you will see them regularly.

```python
# Another common example: classifying sentiment scores
df['sentiment'].apply(lambda x: "positive" if x > 0 else "negative")
```

## 3 Try/Except

A `try/except` block handles errors gracefully. Instead of crashing, Python tries one thing and falls back to another if it fails:

```python
try:
    result = int(value)
except:
    result = 0
```

**How to read it:** "Try to convert `value` to an integer. If that causes an error, set `result` to `0` instead."

This is common in data cleaning when some values in a column cannot be converted — rather than crashing the whole script, `try/except` lets the code handle bad data and keep running.
