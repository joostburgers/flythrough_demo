### 1.2 What Are Variables?

Adding specific numbers like `2 + 1` is fine, but the real magic happens when we use **variables**.

**Think of a variable like a labeled box** where you can store information. 

In the example below:
- We put the number `2` in a box labeled `x`  
- Then we can use `x` instead of writing `2` again
- `x + 1` gives us the same result as `2 + 1`

**Why is this useful?** You'll see in a moment! 🎯

````python
x = 2
x + 1
````

> 👉 **Note:** *In Python, `=` does NOT mean "equals" like in math class. It means "store" or "assign."*

### 1.3 Why Variables Matter

It may not be entirely obvious why you would want to store things in a variable. If you want to add `2` + `1`, storing `2` in the variable `x` seems like an extra step. The reason for doing so is that you do not always know what the information is going to be.

**Example: Calculating a Reddit Post Score**

In later lessons you will be working with Reddit data. Reddit calculates a post's **score** by subtracting downvotes from upvotes:

```
score = upvotes - downvotes
```

Without variables, you would hardcode every number directly — and repeat the formula each time:

```python
1204 - 37    # Post 1
892 - 15     # Post 2
3041 - 128   # Post 3
```

With variables, you store the values once and the formula stays the same:

```python
upvotes = 1204
downvotes = 37
score = upvotes - downvotes
```

Now if the numbers change, you update one value — not the formula. When you get to processing thousands of Reddit posts, Python will swap in new values for `upvotes` and `downvotes` automatically for each one.

## 2 Data Types

### 2.1 Python's Core Data Types

Python can store different **types** of information in variables. Here are the most important types:

1. **String** (`str`) — Text, always wrapped in quotes: `"Hello"`, `"James Madison University"`
2. **Integer** (`int`) — Whole numbers: `1`, `42`, `-5`
3. **List** — An ordered collection of values: `["JMU", "UVA", "Virginia Tech"]`

Two additional types you will encounter:
- **Float** (`float`) — Numbers with decimals: `3.14`, `2.0`
- **Boolean** (`bool`) — True or False values, used in logic

> 👉 **Note:** In Python it is possible to write define a string through both single quotes (`''`) or double quotes (`""`). That is, `'Hello'` and `"Hello"` are both strings and Python parses them the same. The rule is generally to pick one format and stick with it. These lessons generally use double quotes, (`""`), but occasionally use single quotes.

Python automatically figures out what type of data you're storing. This makes Python easier to use than many other programming languages!

````python
# Integer no decimal
a = 1

#float decimal
b = 2.0

#String quote marks
c = "Three"

#Bool reserved word True or False
d = True
````

### 2.3 TypeErrors

The most common mistake beginners make is a `TypeError`. This happens when you try to mix incompatible data types. Some data types can be added and subtracted from one another. For example, `float` and `integer` are both numbers so addition and subtraction is logical. Run the code below to see what happens when `float` and `integer` are added.

✅ **This works:** I can add `a` and `b` because they are both numbers.

````python
a = 1
b = 2.0

a + b
````

> 👉 **Note:** *The result is `3.0` because as soon as an `integer` is combined with a `float` it becomes a `float`.*

**String and Number Combinations**

If I try to add `a` + `c` it will fail because they are incompatible types: numbers and characters. Running this code will result in a `TypeError` for the output.

````python
a = 1
c = "Three"

a + c
````

You should see an error that looks like this:

```python
TypeError: unsupported operand type(s) for +: 'int' and 'str'
```

**What does this mean?** 
- Python is saying: "I can't add an integer (whole number) and a string (text)"

This seems odd. If you see a piece of text that says what is 1 + three, you could easily answer 4 or four. The issue is that a computer fundamentally does not know what a number and a character are. The only thing it knows is that the humans who created Python told it that number and character variables can't be added to one another. Of course, if you ask an AI chatbot to add 1 + Three it will answer four, but not because it knows what those things are but only because it has inferred this from the context. 

**Why this matters:** 
- This isn't a silly mistake! When working with real data, numbers sometimes get saved as text by accident
- Always check your data types when you get strange errors
- Imagine you have a spreadsheet where some numbers are saved as text. Your calculations won't work until you fix the data types!
- If you accidentally enter a variable as a character or a number you will also get this error

**Here's a tricky example:**

Run the cell below and see what happens:

````python
text1 = "Maroon"
text2 = "5"
text1 + text2
````

**Surprise!** This actually works! ✅ 

When you "add" two strings together, you get **string concatenation** (fancy word for "joining text together"). 

This does not work the other way though. Run the line below:

````python
text1 = "Maroon"
text2 = "5"
text1 - text2
````

We get the following error: 

```python
---------------------------------------------------------------------------
TypeError                                 Traceback (most recent call last)
Cell In[3], line 1
----> 1 text1 - text2

TypeError: unsupported operand type(s) for -: 'str' and 'str'
```

**What does this error mean in English?**
- Python is telling us that the `-` (subtraction) operator doesn't work between two strings (`'str'` and `'str'`)
- Unlike `+` which concatenates strings, there's no built-in way to "subtract" text

**How to get help:**
- Read the error message carefully - it tells you exactly what went wrong
- The error type (`TypeError`) indicates a data type problem
- Look at what types are involved (`'str' and 'str'`) and what operation failed (`-`)
- When in doubt, drop the error in the chat, and AI will give you an explanation and possibly a solution in more readable language.

**Numbers vs "Numbers"**

Sometimes numbers look like numbers, but they're actually text! Can you spot the problem below?

````python
number1 = 567
number2 = "5309"
number1 + number2
````

**Problem Alert!**

This gives us a TypeError because:
- `567` is a number (integer)  
- `"5309"` is text (string) - notice the quotes!

Even though "5309" looks like a number, the quotes make Python treat it as text. You can't do math with text!

### 2.4 Type Conversion

**Good news!** You can convert between data types using built-in functions:

- `int()` → Convert to whole number: `int("5309")` → `5309`
- `float()` → Convert to decimal: `float("3.14")` → `3.14`
- `str()` → Convert to text: `str(567)` → `"567"`
- `bool()` → Convert to True/False

**How to use:** Put your variable inside the parentheses: `int("5309")`

**Now our math works!** ✅ By converting the string to an integer with `int()`, we can add them:

```python
number1 + int(number2)   # int() converts "5309" → 5309
```

````python
number1 = 567
number2 = "5309"
number1 + int(number2)
````

## 3 Complex Data Types - Lists and Dictionaries

So far, every variable you have seen holds a single value: a number, a piece of text, or a True/False flag. Python also has **complex data types** that hold multiple values at once. Two of the most common are:

- **List** — an ordered sequence of values, accessed by position
- **Dictionary** — a collection of named values, accessed by key

Both types are fundamental to data science. In Lesson 3 you will work with DataFrames, which are built on top of these structures.

### 3.1 Creating a List

A **list** is a collection of data placed in a specific order. Lists are one of the most common data structures in data science — columns in a data table, rows of results, and search hits are all returned as lists. We will use advanced versions of lists in the upcoming lessons.

You create a list using square brackets, with values separated by commas:

```python
universities = ["JMU", "UVA", "Virginia Tech", "William and Mary"]
```

Lists can hold any type of data: strings, numbers, or even a mix.

> 👉 **Note:** *In other programming languages like JavaScript and C++ these data structures are usually referred to as `array`, but in Python an array is a special data type.*

### 3.2 Accessing List Elements

You can access individual values in a list by using its index. The index always starts at `0`, so the last index value for this list is `3` even though there are four items. We access this value by giving the name of the list and the index in brackets.

````python
list1 = ["JMU", "UVA", "Virginia Tech", "William and Mary"]
````

What value will appear if I type `list1[2]`?

````python
list1[2]
````

#### Negative Index
You can also traverse lists from right to left by putting a `-` in front of the index. This means that `-1` is the last term in the list.

### 3.4 Creating a Dictionary

A **dictionary** stores values under named **keys** instead of numbered positions. Where a list uses an index like `[2]`, a dictionary uses a descriptive label like `['score']`. This makes dictionaries ideal for representing a single structured record, like one Reddit post.

You create a dictionary using curly braces `{}`, with each entry written as `key: value` and separated by commas:

```python
post = {
    'title': 'What dining hall is the best?',
    'score': 142,
    'upvotes': 156,
    'downvotes': 14,
}
```

The key is always a string; the value can be any data type — a string, an integer, a float, or even a list.

````python
post = {
    "title": "What dining hall is the best?",
    "score": 142,
    "upvotes": 156,
    "downvotes": 14,
}

post
````

### 3.5 Accessing Dictionary Values

You access a value from a dictionary by writing the dictionary name followed by the key in square brackets:

```python
post['score']    # retrieves 142
```

This is similar to list indexing, but instead of a number you use the key name. If the key does not exist, Python raises a `KeyError` — similar to the `IndexError` you get when you go out of bounds on a list.

````python
post = {
    "title": "What dining hall is the best?",
    "score": 142,
    "upvotes": 156,
    "downvotes": 14,
}

post["score"]
````

> 📊 **Output:** `142` — Python looked up the value stored under the key `'score'` and returned it directly. The other keys and values in the dictionary were not affected.
