## 1 Basic Styling Principles Adding and Modifying Values in the Function

The cells below rebuild the posts-per-month line chart from Lesson 3.2. In the activities that follow you will practice three targeted modifications: adding markers, switching color swatches, and applying a custom color sequence.

````python
# Import required libraries
import pandas as pd
import plotly.express as px

# Load the pre-cleaned data
df = pd.read_pickle("../data/JMU/JMU_raw_cleaned.pickle")
````

As we learned in the previous lesson, creating a plotly chart is really a matter of running a function and filling in a series of values. So the line:

```python
fig = px.line()
```
Takes the result of `px.line()` stores it into the variable `fig`. Then you can use the method `.show()` to show the Plotly chart.

```python
fig.show()
```

In order to run, the function will always need some basic parameters: data, x-axis, and y-axis. For everything else, Plotly always falls back to a sensible default. You can make as many modifications to any chart as you like, but there are certain principles that are important to remember:

- Parameters always expect a specific variable type. You cannot enter text if numbers are expected
- Always add a comma at the end of the new parameter
- Some parameters have limited set of possible values while others are open

### Markers

The code below recreates the first chart from [lesson 3.2](./lesson_3_2_plotly_charts.ipynb) a line chart showing comments/posts in the Reddit data. By default, Plotly does not add markers to the line chart because this has not been set as an option. 

````python

df["interval"] = df["date"].dt.to_period("M")

# Group the data by interval and post type, then count entries in each group
interval_counts = df.groupby(["interval", "type"], observed=True).size().reset_index(name="count")

# Convert interval back to datetime format so Plotly can understand it
interval_counts["interval"] = interval_counts["interval"].dt.to_timestamp()

# Create the original line chart
fig = px.line(interval_counts, 
              x="interval", 
              y="count", 
              color="type",
              title="Original: Posts and Comments per Month on r/JMU",
              labels={"count": "Number of Posts/Comments", "interval": "Date"})

fig.show()
````

#### Adding Markers

Line charts become easier to read when individual data points are marked. We can add markers to the chart above by passing `markers=True` to `px.line()`. 

````python
fig_modified = px.line(
    interval_counts,
    x="interval",
    y="count",
    color="type",
    title="YOUR TITLE HERE",
    labels={"count": "Your Y-Label", "interval": "Your X-Label"},
    markers=True,
)

fig_modified.show()
````

Set the value to `False` and see what happens.

````python
fig_markers = px.line(
    interval_counts,
    x="interval",
    y="count",
    color="type",
    title="Activity by Month on r/JMU",
    labels={"count": "Posts/Comments", "interval": "Month"},
    markers=True, # Change this to False
       
)
fig_markers.show()
````

> 📊 **Output:** The chart looks identical to the original: a line chart without markers. Setting `markers=False` explicitly produces the same result as omitting it entirely. There is no practical advantage to doing this. In fact, it tends to add clutter. The point is that every chart has many defaults already set, and each one can be overridden to make a chart truly custom.

## 2 Changing Chart Representation

Part 1 covered basic styling options for changing colors and lines, but Plotly also has more sophisticated options to substantially change the representation of the chart. Recall that in [lesson 3.2](./lesson_3_2_plotly_charts.ipynb) we created keyword charts. These were all based on the engagement with a post that contained a particular keyword. It did not distinguish between posts and comments. We can get more insight into the data by plotting keywords and their presence in comments and posts. This can be achieved by setting the `barmode` parameter, which changes how multiple categories are displayed when you pass `color=`.

- `barmode='stack'` - stacks the bars on top of each other. The total height is the combined value across categories.
- `barmode='group'` - places the bars side by side for direct comparison.

These two modes answer different questions. Stacked bars show total volume and composition at the same time. Grouped bars make it easier to compare individual categories directly. The chart below shows the data post/comment data grouped by keyword.

````python

# Count how many posts and comments mention each keyword
keywords = ["tuition", "covid", "party", "football", "class", "library", "campus"]


# Create a helper function to check if text contains a keyword (case-insensitive)
def contains_keyword(text, keyword):
    # Handle missing/empty text
    if pd.isna(text):
        return False
    # Convert both to lowercase to make the search case-insensitive
    return keyword.lower() in text.lower()


keyword_type_counts = []
for keyword in keywords:
    mask = df["text"].apply(lambda x: contains_keyword(x, keyword))
    for post_type in ["comment", "post"]:
        count = df[mask & (df["type"] == post_type)].shape[0]
        keyword_type_counts.append(
            {"keyword": keyword, "type": post_type, "count": count}
        )

keyword_type_df = pd.DataFrame(keyword_type_counts)

# YOUR TURN: Create a grouped bar chart — set barmode='group'
fig_grouped = px.bar(
    keyword_type_df,
    x="keyword",
    y="count",
    color="type",
    barmode="group",  # 'group' places bars side by side
    title="YOUR TITLE HERE",
    labels={"count": "Number of Posts/Comments", "keyword": "Keyword", "type": "Type"},
)

fig_grouped.show()
````

> 📊 **Output:** The keywords are all split between comment/post. Notably some posts elicit many responses like "class". While others like "football" do not get the same response. Meanwhile, the library actually has a really high response ratio. Any time anyone posts about the library people respond.

### ✍️ 2.1 Critical Activity - Posts vs. Comments by Keyword

Now try `barmode='stack'`. Instead of average score, this chart counts how many posts and comments mention each keyword. Placing them one on top the other, it is easier to compare response ratios.


````python
# Count how many posts and comments mention each keyword
keyword_type_counts = []
for keyword in keywords:
    mask = df["text"].apply(lambda x: contains_keyword(x, keyword))
    for post_type in ["comment", "post"]:
        count = df[mask & (df["type"] == post_type)].shape[0]
        keyword_type_counts.append(
            {"keyword": keyword, "type": post_type, "count": count}
        )

keyword_type_df = pd.DataFrame(keyword_type_counts)

fig_stacked = px.bar(
    keyword_type_df,
    x="keyword",
    y="count",
    color="type",
    barmode="group",  # change this so that the bars are stacked
    title="YOUR TITLE HERE",
    labels={"count": "Number of Posts/Comments", "keyword": "Keyword", "type": "Type"},
)

fig_stacked.show()
````

## 3 Styling Chart Layout

So far we have modified charts by changing the options in the main function call `px.bar()`. Generally, the most common data related features are found in this function call. There are also a whole host of modifications that can made of the *presentation* using the method `.update_layout()`. This is always called after the chart has already been created, and generally the features that can be updated through the layout are bit more custom and less common.



### Category Order

The keyword charts in [Lesson 3.2](./lesson_3_2_plotly_charts.ipynb) showed keywords in their original order.

Category order refers to the order in which categories appear on the x or y axis. If the order is not specified, the default is to order them in the way they appear in the list.

```python
keywords = ["tuition", "covid", "party", "football", "class", "library", "campus"]
```

An obvious way to fix this is to simply put them in the order you want them to in the list. This is not a great approach, however, because you are manually doing work a computer could do. Besides, if you decide to add a keyword, you will have to modify the list again. 

On the x-axis you can pass in a dictionary `{}`. A dictionary expects one or more `"key": value` pairs. In this case the key will be `"categoryorder"` and the value can be any one of four: 

- `"category ascending"`
- `"category descending"`
- `"total ascending"`
- `"total descending"`

```python
fig.update_layout(
    xaxis={
        "categoryorder": "total ascending",
    }
)
```
By setting `"categoryorder"` to `"total ascending"` Plotly sorts the values on the x-axis by the values on the y-axis to make for easier comparison. 

````python
# Recreate the keyword analysis from Lesson 3
keywords = ["tuition", "covid", "party", "football", "class", "library", "campus"]

# Helper function to check if text contains a keyword
def contains_keyword(text, keyword):
    if pd.isna(text):
        return False
    return keyword.lower() in text.lower()

# Calculate average scores for each keyword
keyword_scores = []
for keyword in keywords:
    has_keyword_mask = df["text"].apply(lambda x: contains_keyword(x, keyword))
    avg_score = df[has_keyword_mask]["score"].mean()
    keyword_scores.append({"keyword": keyword, "score": avg_score})

keyword_df = pd.DataFrame(keyword_scores)

# Step 1: Create the chart
fig = px.bar(
    keyword_df,
    x="keyword",
    y="score",
    title="Plotly Sorting: Keywords by Score Value",
    labels={
        "keyword": "Keywords",
        "score": "Average Score",
    },
)

# Step 2: Sort the x-axis by the y-values (scores)
fig.update_layout(
    xaxis={
        "categoryorder": "total ascending",  # This sorts by the y-values!
    }
)

fig.show()
````

> 📊 **Output:** It is now much easier to compare between categories. Campus, football, and covid all have similar score, but covid scores the highest.

### Text Case
In the chart above all the category labels are in lower case. While this is not necessarily wrong in all instances, it looks a bit sloppy here. We can update the `xaxis` layout and add the key: `tickfont_textcase`. We can then set the specific way we want text to appear. Our options are:

- `"normal"`
- `"word caps"` 
- `"upper"` 
- `"lower"`

Changing the representation on the axis prevents us from having to change the underlying data.

### Creating a Custom Template

Every chart in this lesson has been passed the same styling arguments: `template=`, `color_discrete_sequence=`, font choices. A **custom template** lets you package all of those decisions once and reuse them across every chart with a single name. This is incredibly powerful, especially if you are creating a website or document and want every chart to have the same look.

Templates can be made by entering custom values for a whole range of properties. Any property that is not explicitly defined will fall back to the default. The template requires several lines of code to work. First, the template definition and then the template registration. Once it is registered, your template can be called inside any function call using `template='my_template'`. You can name your template anything you want, but `my_template` is usually not the best name!

The key properties you can set inside `go.Layout()`:

| Property | Type | What it controls | Example values |
|---|---|---|---|
| `paper_bgcolor` | string | Background color of the canvas outside the plot area | `'white'`, `'#f9f9f9'`, `'rgba(0,0,0,0)'` |
| `plot_bgcolor` | string | Background color inside the axis area | `'#f0f4f8'`, `'white'`, `'rgba(0,0,0,0)'` |
| `font` | `{}` | Default font for all text. Keys: `'family'` (str), `'size'` (int), `'color'` (str) | `{'family': 'Arial', 'size': 12, 'color': '#333333'}` |
| `colorway` | list of strings | Ordered color palette; Plotly cycles through these automatically for each series | `['#e63946', '#457b9d', '#2a9d8f']` |
| `title_font` | `{}` | Font for the chart title only. Same keys as `font` | `{'family': 'Times New Roman', 'size': 16, 'color': '#222222'}` |
| `xaxis` / `yaxis` | `{}` | Axis grid line and tick font settings. Keys: `'showgrid'` (bool), `'gridcolor'` (str), `'tickfont'` (nested `{}` with `'family'`, `'size'`, `'color'`) | `{'showgrid': True, 'gridcolor': '#e0e0e0', 'tickfont': {'family': 'Arial', 'size': 10, 'color': '#555555'}}` |
| `legend` | `{}` | Legend layout and position. Keys: `'orientation'` (`'h'` or `'v'`), `'x'` (float), `'y'` (float) | `{'orientation': 'h', 'x': 0, 'y': -0.2}` |

> 👉 **Note:** *Colors can be written three ways: a named color (`'white'`, `'steelblue'`), a hex code (`'#e63946'`), or an RGBA string (`'rgba(230, 57, 70, 0.5)'`) where the last number controls transparency (0.0 = invisible, 1.0 = fully solid). For `legend`, `x` and `y` are fractions of the chart area: `x=0` is the left edge, `x=1` is the right, and `y=-0.2` places the legend just below the bottom of the plot.*

## 4 Annotating Charts

A good visualization tells a clear story with the minimum number of visual elements necessary. You always want to avoid clutter. Still, sometimes it is useful to guide the user towards a particular data point or to highlight something specific about the chart. This can be done with annotations. As these are essentially ornamentations on an existing chart, the methods always need to be called after the figure is created.



### Vertical Line

You can use a vertical line to demarcate an important moment in the data. 

**`fig.add_vline(x=...)`** — draws a vertical line at a specific x position. For time-series charts, pass a date string like `'2020-08-01'`. Use `line_dash=` to control the line style (`'solid'`, `'dot'`, `'dash'`) and `line_color=` for the color.



#### ✍️ 4.1 Critical Activity - Modifying a vertical line

The code below recreates the keywords over time chart from lesson 3.2. It draws a vertical line at the start of the semester in August 2020, which is when students are most concerned with covid. Try to identify another key moment in the data and move the line there by changing the date.

> 👉 **Note:** by default the charts use the template you created above.

````python

# ── Rebuild keywords over time data (same method as Lesson 3.2) ───────────
KEYWORDS_OVER_TIME = ["tuition", "covid", "football"]
INTERVAL = "M"
MIN_PERIOD_POSTS = 30

_df_trend = df.copy()
_df_trend["interval"] = _df_trend["date"].dt.to_period(INTERVAL).dt.to_timestamp()

_period_totals = (
    _df_trend.groupby("interval")
             .size()
             .reset_index(name="_total")
)
_period_totals = _period_totals[_period_totals["_total"] >= MIN_PERIOD_POSTS]

_rows = []
for kw in KEYWORDS_OVER_TIME:
    _mask = _df_trend["text"].str.contains(kw, case=False, na=False)
    _hits = (
        _df_trend[_mask]
        .groupby("interval")
        .size()
        .reset_index(name="_hits")
    )
    _m = _hits.merge(_period_totals, on="interval")
    _m["per_1000"] = _m["_hits"] / _m["_total"] * 1000
    _m["keyword"] = kw
    _rows.append(_m[["interval", "keyword", "per_1000"]])

_keyword_trend_df = pd.concat(_rows, ignore_index=True)

# ── Build the base chart ──────────────────────────────────────────────────
fig_annotated = px.line(
    _keyword_trend_df,
    x="interval",
    y="per_1000",
    color="keyword",
    title="Keyword Mentions Over Time on r/JMU",
    labels={
        "interval": "Time Period",
        "per_1000": "Mentions per 1,000 Posts/Comments",
        "keyword": "Keyword",
    },
    markers=True,
)

# ── Add a vertical line at August 2020 ────────────────────────────────────
# line_dash options: 'solid', 'dot', 'dash', 'longdash', 'dashdot'
fig_annotated.add_vline(
    x="2020-08-01",
    line_dash="dash",
    line_color="red",
    line_width=1,
)

fig_annotated.show()
````

### Adding text annotations

You can also add text to the chart directly. This is a bit trickier, because you have to pick a specific point on the grid to place the arrow and the text. There are some basic rules for doing so, but this can be mostly a lot of visual checking and making sure something looks good.

`fig.add_annotation()` places a labeled callout with an optional arrow at a specific chart coordinate:

| Parameter | Type | Description | Example |
|---|---|---|---|
| `x` | string or number | X-axis position of the annotation point. For time-series, pass a date string | `'2020-08-01'` |
| `y` | number | Y-axis value where the annotation points | `45.2` or `_df['col'].max() * 0.9` |
| `text` | string | Label text displayed near the arrow | `'COVID spike'` |
| `showarrow` | bool | Whether to draw an arrow from the label to the point | `True` |
| `arrowhead` | int (0–8) | Arrow tip style. `0` = no tip, `2` = filled triangle, `6` = open arrow | `2` |
| `ax` | number (pixels) | Horizontal offset of the label from the annotation point. Positive moves right | `50` |
| `ay` | number (pixels) | Vertical offset of the label from the annotation point. Negative moves up | `-40` |
| `font` | `dict()` | Font for the label text. Keys: `size` (int), `color` (str) | `dict(size=12, color='red')` |


> 👉 **Note:** *`ax` and `ay` are pixel offsets, not data coordinates. `ax=50, ay=-40` places the label 50 pixels to the right and 40 pixels above the annotation point. Adjust these until the label clears the data line.*

The code below marks the August 2020 covid spike with a vertical line and a text callout. Move both to a different moment in the data and update the annotation text to match.

**Tip:** Hover over any data point in the chart above to read its exact date and `per_1000` value. Copy those directly into `x=` and `y=` to place your annotation precisely.

````python
# ── Rebuild keywords over time data (same method as Lesson 3.2) ───────────
KEYWORDS_OVER_TIME = ["tuition", "covid", "football"]
INTERVAL = "M"
MIN_PERIOD_POSTS = 30

_df_trend = df.copy()
_df_trend["interval"] = _df_trend["date"].dt.to_period(INTERVAL).dt.to_timestamp()

_period_totals = (
    _df_trend.groupby("interval")
             .size()
             .reset_index(name="_total")
)
_period_totals = _period_totals[_period_totals["_total"] >= MIN_PERIOD_POSTS]

_rows = []
for kw in KEYWORDS_OVER_TIME:
    _mask = _df_trend["text"].str.contains(kw, case=False, na=False)
    _hits = (
        _df_trend[_mask]
        .groupby("interval")
        .size()
        .reset_index(name="_hits")
    )
    _m = _hits.merge(_period_totals, on="interval")
    _m["per_1000"] = _m["_hits"] / _m["_total"] * 1000
    _m["keyword"] = kw
    _rows.append(_m[["interval", "keyword", "per_1000"]])

_keyword_trend_df = pd.concat(_rows, ignore_index=True)

# ── Build the base chart ──────────────────────────────────────────────────
fig_annotated = px.line(
    _keyword_trend_df,
    x="interval",
    y="per_1000",
    color="keyword",
    title="Keyword Mentions Over Time on r/JMU",
    labels={
        "interval": "Time Period",
        "per_1000": "Mentions per 1,000 Posts/Comments",
        "keyword": "Keyword",
    },
    markers=True,
)

# ── Add a vertical line at August 2020 ────────────────────────────────────
# line_dash options: 'solid', 'dot', 'dash', 'longdash', 'dashdot'
fig_annotated.add_vline(
    x="2020-08-01",
    line_dash="dash",
    line_color="red",
    line_width=1,
)

# ── Add a text callout near the spike ─────────────────────────────────────
# ax/ay control the pixel offset of the arrow tail from the label point
fig_annotated.add_annotation(
    x="2020-08-01",
    y= 57,
    text="Start of Fall 2020 Semester",
    showarrow=True,
    arrowhead=2,
    ax=-80,
    ay=-60,
)

fig_annotated.update_layout(
    xaxis_title="Time Period",
    yaxis_title="Mentions per 1,000 Posts/Comments",
)
fig_annotated.show()
````

### When Annotations Go Wrong

You can call `add_vline()` and `add_annotation()` as many times as you like. Each call simply appends another marking to the existing figure. This makes it tempting to label every interesting moment you spot. The chart below does exactly that.

Run it and look carefully at the result before reading the reflection below.

````python
fig_cluttered = px.line(
    _keyword_trend_df,
    x="interval",
    y="per_1000",
    color="keyword",
    title="Every Interesting Moment (Please Do Not Do This)",
    labels={
        "interval": "Time Period",
        "per_1000": "Mentions per 1,000 Posts/Comments",
        "keyword": "Keyword",
    },
    markers=True,
    template="my_template",
)

events = [
    ("2019-08-01", 10, "Fall 2019 semester start",    50,  -40),
    ("2020-03-01", 20, "COVID declared pandemic",     -120, -40),
    ("2020-08-01", 57, "Fall 2020 semester (remote)", -130, -30),
    ("2021-01-01", 15, "Spring 2021 return",           50,  -40),
    ("2021-08-01", 12, "Fall 2021 in-person return",   50,  -50),
    ("2022-01-01",  8, "Spring 2022",                  50,  -40),
]

for date, y_val, label, ax, ay in events:
    fig_cluttered.add_vline(
        x=date,
        line_dash="dot",
        line_color="gray",
        line_width=1,
    )
    fig_cluttered.add_annotation(
        x=date,
        y=y_val,
        text=label,
        showarrow=True,
        arrowhead=2,
        ax=ax,
        ay=ay,
    )

fig_cluttered.show()
````

> 💡 **Reflection:** Six lines and six labels, and the chart is now harder to read than the version with none. Labels overlap, arrows cross the data, and the eye has nowhere to land. Generally, you should limit yourself to *one* annotation. On rare occasions, you might consider *two* annotations if you are showing a contrast.

## Lesson Summary

### Part 1: Line Chart Styling

- `markers=True` — adds dot markers to a `px.line()` chart
- `color_discrete_sequence=` — overrides the default color palette for a single chart
- `px.colors.qualitative.swatches()` — displays all available named color swatches
- `px.colors.qualitative.G10` — an example named color sequence (replace with any swatch name)

### Part 2: Changing Chart Representation

- `barmode="stack"` - stacks bars on top of each other; shows total volume and composition together
- `barmode="group"` - places bars side by side; makes it easier to compare categories directly
- Both are passed as a parameter to `px.bar()`

### Part 3: Layout Tweaks

- `fig.update_layout(xaxis={'categoryorder': '...'})` - sorts bars by the order you specify
- `'tickfont_textcase': 'word caps'` - capitalizes tick labels without changing the underlying data
- `template=` - applies a named style to a chart (e.g. `'ggplot2'`, `'seaborn'`, `'plotly_dark'`)
- `go.layout.Template()` — creates a reusable template object
- `pio.templates["my_template"] = my_template` — registers the template by name

### Part 4: Annotations

- `fig.add_vline(x="2020-08-01", line_dash="dot", line_color="red")` — draws a vertical reference line
- `fig.add_annotation(x=..., y=..., text=..., ax=..., ay=...)` — places a labeled callout with an arrow
- Pass `template="my_template"` to any `px` function to apply the custom style
