## 📖 2 Follow Along — Apply Styles

You do not need to write or modify any code in this section. Run each cell and focus on understanding what the code is doing and why.

The cell below:
1. Writes `team_theme.css` with your chosen colors and font
2. Adds a `<link>` tag to each HTML page so they all load the new stylesheet (safe to re-run — it only adds the tag once)
3. Updates the banner subtitle on every page if you provided one

Click the <img src="../lesson_assets/images/jupyter/run.svg" style="height: 14pt; vertical-align: middle;"> button below to apply your style choices.

```python
import os
import re
import json

if "STYLE_CONFIG" not in dir():
    print("⛔ Run the STYLE_CONFIG cell above first.")
else:
    _HTML_FILES = ["index.html", "team.html", "whitepaper.html", "interactive_tour.html"]

    # ── Google Font lookup table ──────────────────────────────────────────────────
    _GOOGLE_FONTS = {
        "Lato":             "Lato:wght@300;400;700",
        "Open Sans":        "Open+Sans:wght@300;400;600;700",
        "Merriweather":     "Merriweather:wght@300;400;700",
        "Montserrat":       "Montserrat:wght@300;400;600;700",
        "Raleway":          "Raleway:wght@300;400;600;700",
        "Roboto":           "Roboto:wght@300;400;500;700",
        "Nunito":           "Nunito:wght@300;400;600;700",
        "Playfair Display": "Playfair+Display:wght@400;600;700",
        "PT Serif":         "PT+Serif:wght@400;700",
    }

    _font = STYLE_CONFIG["font_family"].strip()

    _import_line = ""
    for _name, _spec in _GOOGLE_FONTS.items():
        if _name in _font:
            _import_line = f"@import url('https://fonts.googleapis.com/css2?family={_spec}&display=swap');\n\n"
            if " " in _name and not _font.startswith("'") and not _font.startswith('"'):
                _font = f"'{_name}', sans-serif"
            break

    # ── 1. Write team_theme.css (site-wide: colors + font) ───────────────────────
    _css = (
        _import_line +
        "/* Team theme — auto-generated by project_part_4_global_variables.ipynb */\n"
        "/* Do not edit by hand. Re-run the apply cell in Part 4 to update.       */\n\n"
        ":root {\n"
        f"    --primary-color:   {STYLE_CONFIG["primary_color"]};\n"
        f"    --secondary-color: {STYLE_CONFIG["secondary_color"]};\n"
        "}\n\n"
        "body {\n"
        f"    font-family: {_font};\n"
        "}\n\n"
        "h1, h2, h3, h4, h5, h6 {\n"
        f"    font-family: {_font};\n"
        "}\n"
    )
    with open("team_theme.css", "w", encoding="utf-8") as _f:
        _f.write(_css)
    print("✅ team_theme.css written")

    # ── 2. Write interactive_tour_theme.css (tour-specific overrides) ─────────────
    _counter_display = "block" if STYLE_CONFIG.get("show_chapter_counter", True) else "none"
    _flex_dir        = "row-reverse" if STYLE_CONFIG.get("map_side", "right") == "left" else "row"

    _it_css = (
        "/* Interactive Tour theme — auto-generated by project_part_4_global_variables.ipynb */\n"
        "/* Do not edit by hand. Re-run the apply cell in Part 4 to update.                  */\n\n"
        ":root {\n"
        f"    --tour-accent: {STYLE_CONFIG.get("accent_color", "#4a6fa5")};\n"
        "}\n\n"
        "article {\n"
        f"    max-width: {STYLE_CONFIG.get("card_max_width", "40rem")};\n"
        "}\n\n"
        ".step {\n"
        f"    background-color: {STYLE_CONFIG.get("card_background", "rgba(255, 255, 255, 0.95)")};\n"
        f"    padding:          {STYLE_CONFIG.get("card_padding", "25px 50px")};\n"
        f"    opacity:          {STYLE_CONFIG.get("card_opacity_inactive", 0.5)};\n"
        "}\n\n"
        ".step.is-active {\n"
        "    opacity: 1;\n"
        "}\n\n"
        ".step-title {\n"
        f"    font-size:  {STYLE_CONFIG.get("font_size_title", "24px")};\n"
        f"    color:      {STYLE_CONFIG.get("title_color", "#2d3748")};\n"
        f"    text-align: {STYLE_CONFIG.get("title_alignment", "left")};\n"
        "}\n\n"
        ".step p {\n"
        f"    font-size: {STYLE_CONFIG.get("font_size_body", "16px")};\n"
        f"    color:     {STYLE_CONFIG.get("body_text_color", "#444")};\n"
        "}\n\n"
        ".step blockquote {\n"
        "    border-left-color: var(--tour-accent);\n"
        "    background: color-mix(in oklab, var(--tour-accent) 8%, white);\n"
        "}\n\n"
        ".step-badge-combined-date,\n"
        ".date-range-badge {\n"
        "    background: var(--tour-accent);\n"
        "}\n\n"
        ".step-counter {\n"
        f"    display: {_counter_display};\n"
        "}\n\n"
        "#scrolly {\n"
        f"    flex-direction: {_flex_dir};\n"
        "}\n"
    )
    with open("interactive_tour_theme.css", "w", encoding="utf-8") as _f:
        _f.write(_it_css)
    print("✅ interactive_tour_theme.css written")

    # ── 3. Patch HTML files ───────────────────────────────────────────────────────
    _TEAM_LINK = '    <link rel="stylesheet" href="./team_theme.css">'
    _IT_LINK   = '    <link rel="stylesheet" href="./interactive_tour_theme.css">'

    _page_title = STYLE_CONFIG.get("page_title", "").strip()
    _subtitle   = STYLE_CONFIG.get("subtitle",   "").strip()
    _sub_updated = 0

    for _html in _HTML_FILES:
        if not os.path.exists(_html):
            print(f'⚠️  {_html} not found — skipping')
            continue

        with open(_html, "r", encoding="utf-8") as _f:
            _content = _f.read()

        # Inject team_theme.css link after styles.css (idempotent)
        if "team_theme.css" not in _content:
            _content = re.sub(
                r'(<link[^>]+href="\.?/?\.?styles\.css"[^>]*>)',
                r"\1\n" + _TEAM_LINK,
                _content, count=1
            )
            print(f'✅ {_html} — team_theme.css link added')
        else:
            print(f'ℹ️  {_html} — team_theme.css link present')

        # interactive_tour.html: inject tour theme link, patch <title> and banner H1
        if _html == "interactive_tour.html":
            if "interactive_tour_theme.css" not in _content:
                _content = _content.replace("</head>", _IT_LINK + "\n</head>", 1)
                print(f'✅ {_html} — interactive_tour_theme.css link added')
            else:
                print(f'ℹ️  {_html} — interactive_tour_theme.css link present')

            if _page_title:
                _content = re.sub(
                    r"<title>[^<]*</title>",
                    f'<title>{_page_title}</title>',
                    _content, count=1
                )
                _content = re.sub(
                    r'(<div class="banner">.*?<h1>)[^<]*(</h1>)',
                    lambda m: m.group(1) + _page_title + m.group(2),
                    _content, flags=re.DOTALL, count=1
                )
                print(f'✅ {_html} — page title set to "{_page_title}"')

        # Update banner subtitle
        if _subtitle:
            _patched = re.sub(
                r'(<div class="banner">.*?<p>)([^<]*)(</p>)',
                lambda m: m.group(1) + _subtitle + m.group(3),
                _content, flags=re.DOTALL
            )
            if _patched != _content:
                _content = _patched
                _sub_updated += 1

        with open(_html, "w", encoding="utf-8") as _f:
            _f.write(_content)

    if _subtitle and _sub_updated:
        print(f'✅ Banner subtitle updated in {_sub_updated} file(s): "{_subtitle}"')
    elif _subtitle:
        print("ℹ️  Banner subtitle — no changes (already matches or pattern not found)")
    else:
        print("ℹ️  No subtitle — banner taglines unchanged")

    # ── 4. Patch footer in interactive_tour_config.js ─────────────────────────────
    _footer_text = STYLE_CONFIG.get("footer_text", "").strip()
    _config_js   = "interactive_tour_config.js"
    if _footer_text:
        if os.path.exists(_config_js):
            with open(_config_js, "r", encoding="utf-8") as _f:
                _js = _f.read()
            _footer_val = json.dumps(_footer_text)[1:-1]  # JSON-encode the inner value only
            _js_new = re.sub(r'"footer":\s*"[^"]*"', f'"footer": "{_footer_val}"', _js, count=1)
            if _js_new != _js:
                with open(_config_js, "w", encoding="utf-8") as _f:
                    _f.write(_js_new)
                print(f'✅ interactive_tour_config.js — footer updated: "{_footer_text}"')
            else:
                print("ℹ️  interactive_tour_config.js — footer key not found; re-run the Part 3 export cell first")
        else:
            print("ℹ️  interactive_tour_config.js not found — run the Part 3 export cell first, then re-run this cell")

    print()
    print("Reload any open HTML pages in your browser to see the changes.")
```
