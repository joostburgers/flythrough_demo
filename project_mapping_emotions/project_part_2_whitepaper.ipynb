### Key Topics

📋 **Instructions:**

**Task:** Provide a data-driven analysis of the most discussed topics in each subreddit.
**Length:** 5–8 sentences
**Detail:** The chart below shows how frequently each keyword appears across posts and comments for both schools. Your first and most important task is to **decide which keywords are worth comparing**. A keyword like `class` will appear in almost every academic subreddit and tell you nothing interesting; a keyword like `tailgate` or `dining hall` may reveal something specific about campus culture at one school.

To find good keywords:
- Skim 20–30 posts from each school's `_raw.csv` file and note recurring topics, complaints, or celebrations
- Think about what is *unique* to each school's context (athletics, geography, campus life, local events)
- Avoid keywords so generic they appear everywhere (`good`, `people`, `year`) or so rare they barely register

Once you have a chart you find meaningful, analyze it: which topics are **shared** across both schools, and which are **distinctive** to one? Do the differences reflect something meaningful about each institution?

> 👉 **Note:** *Because the two corpora are different sizes, the chart uses counts per 1,000 posts so the schools are directly comparable.*


```python

# ── Keywords to compare — change this list ───────────────────────────────
# Tip: skim each school's _raw.csv first to find topics that are actually
# discussed. Generic words ('good', 'people') or rare words produce flat bars.
KEYWORDS = ["tuition", "covid", "party", "football", "basketball", "class", "zoom", "campus"]

# ── Load raw data for both schools ────────────────────────────────────────
_df_jmu    = pd.read_pickle("../data/JMU/JMU_raw_cleaned.pickle")
_df_school = pd.read_pickle(f'../data/{SCHOOL}/{SCHOOL}_raw.pickle')

# ── Count keyword mentions and normalize per 1,000 posts ─────────────────
# Normalizing makes the two corpora directly comparable even though they
# have different numbers of posts and comments.
_jmu_total    = len(_df_jmu)
_school_total = len(_df_school)

_rows = []
for kw in KEYWORDS:
    _jmu_hits    = _df_jmu["text"].str.contains(kw, case=False, na=False).sum()
    _school_hits = _df_school["text"].str.contains(kw, case=False, na=False).sum()
    _rows.append({"keyword": kw, "school": "JMU",    "per_1000": _jmu_hits    / _jmu_total    * 1000})
    _rows.append({"keyword": kw, "school": SCHOOL,   "per_1000": _school_hits / _school_total * 1000})

_keyword_df = pd.DataFrame(_rows)

# ── Plot ──────────────────────────────────────────────────────────────────
fig = px.bar(
    _keyword_df,
    x="keyword",
    y="per_1000",
    color="school",
    barmode="group",
    title=f'Keyword Frequency per 1,000 Posts: JMU vs {SCHOOL}',
    labels={"per_1000": "Mentions per 1,000 Posts/Comments",
            "keyword": "Keyword", "school": "School"},
    template="my_template"
)
fig.update_layout(
    xaxis_title="Keyword",
    yaxis_title="Mentions per 1,000 Posts/Comments",
    legend_title="School",
    xaxis={"categoryorder": "total descending"},
)
fig.show()
```

### Keyword Trends


📋 **Instructions:**

**Task:** Identify up to three keywords that show the most meaningful difference — or the most surprising similarity — between the two schools and plot how their usage evolved over time.
**Length:** 5–8 sentences
**Detail:** Looking at your Key Topics bar chart, identify up to three keywords that have the highest analytical payoff: either a keyword one school uses far more than the other, or one that both schools spike around the same time (suggesting a shared event like COVID or a regional news story). You may use fewer than three if two keywords already tell a complete story — more lines can make the chart harder to read.

To decide which keywords to keep:
- A keyword is worth including if it produces at least one noticeable peak or divergence you can explain
- If two keywords behave nearly identically, drop one — they are telling the same story
- A flat line for both schools is not interesting unless the *absence* of a topic is itself meaningful

**Adjusting the chart:** Two variables control how the chart looks:
- `TREND_INTERVAL` — the time grouping. Finer intervals (`'W'` weekly, `'M'` monthly) show more detail but can produce spiky lines when post volume is low. Coarser intervals (`'Q'` quarterly, `'Y'` yearly) smooth the lines and reduce distortion because each bucket has more posts to normalize against. Start with `'Q'` and zoom in only if a quarterly chart already shows a clear story.
- `MIN_PERIOD_POSTS` — any interval with fewer total posts than this threshold is dropped before normalization. Raise it to remove more noisy periods; lower it to keep them.

**Color** distinguishes the two schools; **line pattern** (solid vs. dashed vs. dotted) distinguishes the keywords. Your written analysis should name each peak, explain what was happening at that time, and say what it reveals about each institution.

**Annotations** help point out important patterns in the data. Two have been created for you. Adjust them as necessary. You may also opt to delete them. See [Lesson 3.3](../lesson_3_introduction_pandas/lesson_3_3_plotly_styling.ipynb) for details.





```python
# ── Pick 1–3 keywords from your Key Topics chart ─────────────────────────
# Choose keywords that produced the most interesting peaks or divergences.
# You can use fewer than 3 if two keywords already tell a complete story.
TREND_KEYWORDS = ["covid", "football", "tuition"]  # ← change these (1–3 keywords)

# ── Time interval for this chart ──────────────────────────────────────────
# Coarser intervals (Q, Y) pool more posts per bucket, which reduces
# normalization distortion and produces smoother lines.
# Options: 'W' (weekly), 'M' (monthly), 'Q' (quarterly), 'Y' (yearly)
TREND_INTERVAL = "Y"  # ← change if needed (default: quarterly)

# ── Minimum posts per interval to include in the chart ───────────────────
# Intervals below this threshold are dropped before normalizing.
# Raise to suppress noisy periods; lower to include them.
MIN_PERIOD_POSTS = 10  # ← change if needed

# ── Load raw data for both schools ────────────────────────────────────────
_df_jmu    = pd.read_pickle("../data/JMU/JMU_raw_cleaned.pickle")
_df_school = pd.read_pickle(f'../data/{SCHOOL}/{SCHOOL}_raw.pickle')

_df_jmu["school"]    = "JMU"
_df_school["school"] = SCHOOL

_df_combined = pd.concat([_df_jmu, _df_school], ignore_index=True)

_df_combined["interval"] = _df_combined["date"].dt.to_period(TREND_INTERVAL).dt.to_timestamp()

# ── Normalize per 1,000 — drop intervals with too few posts ──────────────
# A month with 1 total post and 1 covid hit = 1,000 per 1,000: not meaningful.
_period_totals = (_df_combined.groupby(["interval", "school"])
                               .size()
                               .reset_index(name="_total"))
_period_totals = _period_totals[_period_totals["_total"] >= MIN_PERIOD_POSTS]

_rows = []
for kw in TREND_KEYWORDS:
    _mask = _df_combined["text"].str.contains(kw, case=False, na=False)
    _hits = (_df_combined[_mask]
             .groupby(["interval", "school"])
             .size()
             .reset_index(name="_hits"))
    _m = _hits.merge(_period_totals, on=["interval", "school"])
    _m["per_1000"] = _m["_hits"] / _m["_total"] * 1000
    _m["keyword"] = kw
    _rows.append(_m[["interval", "school", "keyword", "per_1000"]])

_trend_df = pd.concat(_rows, ignore_index=True)

# ── Plot — color = school, line pattern = keyword ─────────────────────────
fig = px.line(
    _trend_df,
    x="interval",
    y="per_1000",
    color="school",
    line_dash="keyword",
    title=f'Keyword Trends Over Time: JMU vs {SCHOOL}',
    labels={"interval": "Time Period", "per_1000": "Mentions per 1,000 Posts/Comments",
            "school": "School", "keyword": "Keyword"},
    markers=True,
    template="my_template"
)
fig.update_layout(
    xaxis_title="Time Period",
    yaxis_title="Mentions per 1,000 Posts/Comments",
)

fig.add_vline(
    x="2020-08-01",
    line_dash="dash",
    line_color="#450084",
    line_width=1,
)


fig.add_annotation(
    x="2020-08-01",
    y=25,
    text="RANDOM ANNOTATION 1",
    showarrow=True,
    arrowhead=2,
    ax=-80,
    ay=-60,
)

fig.add_vline(
    x="2021-08-01",
    line_dash="dash",
    line_color="#4B9CD3",
    line_width=1,
)

fig.add_annotation(
    x="2021-08-01",
    y=25,
    text="RANDOM ANNOTATION 2",
    showarrow=True,
    arrowhead=2,
    ax=80,
    ay=-60,
)

fig.show()
```
